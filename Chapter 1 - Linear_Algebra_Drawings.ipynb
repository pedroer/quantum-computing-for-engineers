```python
# Draw two vectors in a Cartesian grid

# Define vectors
a = [3, 2]
b = [1, -2]

# Create the plot
plt.figure(figsize=(8, 8))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='red',   label='Vector a=[3,  2]')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector b=[1, -2]')

# Add dark X and Y axes with arrows
plt.quiver(-3, 0, 7, 0, angles='xy', scale_units='xy', scale=1, color='gray')  # X-axis
plt.quiver(0, -3, 0, 7, angles='xy', scale_units='xy', scale=1, color='gray')  # Y-axis

# draw the dashed lines to facilitate visualization of vector a
plt.plot([3, 3], [0, 2], 'r--')
plt.plot([0, 3], [2, 2], 'r--')

# draw the dashed lines to facilitate visualization of vector b
plt.plot([0, 1], [-2, -2], 'g--')
plt.plot([1, 1], [0, -2], 'g--')

# Set grid with light alpha
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.5)

# Set limits
plt.xlim(-3, 4)
plt.ylim(-3, 4)

# Label axes
plt.xlabel('X')
plt.ylabel('Y')

# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define vectors
a = [3, 2]
b = [1, -2]

# Create the plot
plt.figure(figsize=(8, 8))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='darkred', label='Vector a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='darkgreen', label='Vector b')

# Add gray X and Y axes with arrows
plt.quiver(-3, 0, 7, 0, angles='xy', scale_units='xy', scale=1, color='gray')  # X-axis
plt.quiver(0, -3, 0, 7, angles='xy', scale_units='xy', scale=1, color='gray')  # Y-axis

# Set grid with light alpha
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.5)

# Set limits
plt.xlim(-3, 4)
plt.ylim(-3, 4)

# Label axes
plt.xlabel('X')
plt.ylabel('Y')

# Add ticks for X and Y axes
plt.xticks(range(-3, 4))
plt.yticks(range(-3, 4))

# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
