```python
#-------------------------------------------------------------------------
# Grover’s Oracle
# Chapter 2 in the QUANTUM COMPUTING AND QUANTUM MACHINE LEARNING BOOK
#-------------------------------------------------------------------------
# Version 1.0
# (c) 2025 Jesse Van Griensven, Roydon Fraser, and Jose Rosas 
# Licence:  MIT - Citation required
#-------------------------------------------------------------------------
import warnings
warnings.filterwarnings('ignore')

from qiskit import QuantumCircuit
from qiskit.visualization import circuit_drawer
#-------------------------------------------------------------------------

qc = QuantumCircuit(2)
qc.h([0, 1])  # Apply Hadamard to create superposition
qc.cz(0, 1)   # Apply Controlled-Z
qc.h([0, 1])  # Apply Hadamard again
qc.measure_all()

# Visualize the circuit
print("Grover’s Oracle Simple Quantum Circuit:")
print(qc.draw())
display(circuit_drawer(qc, output='mpl', style="iqp"))
```
