```python
#-------------------------------------------------------------------------
# QFT for Key Analysis
# Chapter 5 in the QUANTUM COMPUTING AND QUANTUM MACHINE LEARNING BOOK
#-------------------------------------------------------------------------
# Version 1.0
# (c) 2025 Jesse Van Griensven, Roydon Fraser, and Jose Rosas 
# Licence:  MIT - Citation required
#-------------------------------------------------------------------------
import warnings
warnings.filterwarnings('ignore')

from qiskit import Aer, QuantumCircuit, execute
from qiskit.visualization import circuit_drawer

from qiskit.circuit.library import QFT
#-------------------------------------------------------------------------

# Define QFT circuit
n  = 3
qc = QuantumCircuit(n)
qc.append(QFT(num_qubits=n).to_instruction(), range(n))
qc.measure_all()

# Draw the circuit
display(circuit_drawer(qc, output='mpl', style="iqp"))

# Simulate the circuit
simulator = Aer.get_backend('qasm_simulator')
result    = execute(qc, simulator, shots=1024).result()
print("Cryptographic Key Analysis Results:", result.get_counts())
```
